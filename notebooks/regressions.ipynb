```python
import sys
sys.path.append("../src")

import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from functions import *

pd.set_option('display.float_format','{:.3f}'.format)
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
df = pd.read_csv("../data/01_raw/data.csv")
```

```python
get_corr_pairs(df, 10)
```

```python
y = df.SalePrice
feats = [
    "OverallQual",
    "GrLivArea",
    "GarageCars",
    "GarageArea",
    "TotalBsmtSF",
    "1stFlrSF"
]
X=df[feats]

```

```python
Xs = []

for i in range(4):
    Xs.append(df[feats[i:]])

print(Xs)
```

# Régression linéaire

```python
for i in range(len(Xs)):
    print(f"Avec X_{i} = {list(Xs[i].columns)}\n")
    get_model_metrics(LinearRegression(), Xs[i], y)
```

# Régression linéaire Ridge


```python
# Liste des alpha à utiliser pour les Lasso() et les Ridge()
alphas = [200*i+1 for i in range(6)]
```

```python
for i in range(len(Xs)):
    print(f"Avec X_{i} = {list(Xs[i].columns)}\n")
    for a in alphas:
        print(f"Avec α = {a} :")
        get_model_metrics(Ridge(alpha=a), Xs[i], y)
        print("\n")
```

# Régression linéaire Lasso

```python
for i in range(len(Xs)):
    print(f"Avec X_{i} = {list(Xs[i].columns)}\n")
    for a in alphas:
        print(f"Avec α = {a} :")
        get_model_metrics(Lasso(alpha=a), Xs[i], y)
        print("\n")
```
